```python
import numpy as np
import math
from matplotlib import colors
```

```python
""" Initial conditions. """
interest_rate = 0.04
volatility = 0.9
S_0 = 100
strike_price = 110

maturity = 1
delta_t = 0.01


delta_x = 100
Vmax = 60 
V_0=9.6253


M_low = 0
M_high = delta_x

vec_orig = np.zeros(M_high)
for i in range(M_high):
    if i * delta_x > V_0 - 0.5 *delta_x:
        vec_orig[i] = 1
        break

timemat = []

```

```python
"""The simplified terms"""
alpha = (interest_rate - (1/2) * volatility ** 2) * (delta_t / (2 * delta_x))
beta = ((1 / 2) * volatility ** 2) * (delta_t / (delta_x ** 2))
gamma = interest_rate * delta_t
```

```python
""" The FTCS scheme"""
a_prev = -alpha + beta
a_cur = 1 - 2 * beta -gamma
a_next = alpha + beta
b_prev = 0
b_cur = 1
b_next =0
```

```python
""" CHECK EUROPEAN CALL"""
import copy

N = M_high


A = np.zeros((N,N))
B = np.zeros((N,N))

for i in range(N):
    for j in range(N):
            
        if i == j:
            A[i][j] = a_cur
            B[i][j] = b_cur
            if j < N-1:
                A[i][(j + 1)] = a_next
                B[i][(j + 1)] = b_next
            if j > 0:
                A[i][(j - 1)%N] = a_prev          
                B[i][(j - 1)%N] = b_prev

t = 0.01

Vn = vec_orig.copy()
maxVN = sum(Vn)
for i in range(len(Vn)):
    Vn[i] = Vn[i]/maxVN*100
mat_time = [Vn]
for iteration in range(100000):
    Vn = np.dot(A, Vn)
    if iteration % 1000 == 0:
        maxVN = sum(Vn)
        for i in range(len(Vn)):
            Vn[i] = Vn[i]/maxVN*100
        mat_time.append(Vn.copy())
```

## Crank Nicolson

```python
""" The Crank-Nicolson scheme"""
a_prev = (1/2) * -alpha + (1/2) * beta
a_cur = 1 - beta - (1/2) * gamma
a_next = (1/2) * alpha + (1/2) * beta
b_prev = (1/2) * alpha - (1/2) * beta
b_cur = 1 + beta + (1/2) * gamma
b_next = -(1/2) * alpha - (1/2) * beta

```

```python
""" CHECK EUROPEAN CALL"""
import copy

N = M_high


A = np.zeros((N,N))
B = np.zeros((N,N))

for i in range(N):
    for j in range(N):
            
        if i == j:
            A[i][j] = a_cur
            B[i][j] = b_cur
            if j < N-1:
                A[i][(j + 1)] = a_next
                B[i][(j + 1)] = b_next
            if j > 0:
                A[i][(j - 1)%N] = a_prev          
                B[i][(j - 1)%N] = b_prev
        
    
t = 0.01

B_inv = np.linalg.inv(B)

Vn = vec_orig.copy()
maxVN = max(Vn)
for i in range(len(Vn)):
    Vn[i] = Vn[i]/maxVN*100
mat_time = [Vn]
for iteration in range(100000):
    Vn = np.dot(A, Vn)
    Vn = np.dot(B_inv, Vn)
    if iteration % 1000 == 0:
        maxVN = max(Vn)
        for i in range(len(Vn)):
            Vn[i] = Vn[i]/maxVN*100
        mat_time.append(Vn.copy())
```

```python
k = 100

#B Vn+1 = AVn
#A-1 B VT = VT-1

numbergrids = delta_x
Xmin = 90
Xmax = 110


Xtmat = []
for t in np.arange(numbergrids):
    temp = []
    for X in np.linspace(Xmin, Xmax, numbergrids):
        temp.append([X,(numbergrids - t)/numbergrids, 0])
    Xtmat.append(temp)

for el in  Xtmat[0]:
    el[2] = max(el[0] - k, 0)

matrixsol = []    
vec = []

for v in range(len(Xtmat[0])):
    vec.append(Xtmat[0][v][2])
matrixsol.append(np.array(vec))


for _ in range(int(maturity/delta_t)-1):
    #print(np.dot(stepback, matrixsol[-1]))
    matrixsol.append(np.dot(stepback, matrixsol[-1]))
    
X = []
Y = []

for _ in range(int(maturity/delta_t)):    
    Y.append(np.linspace(Xmin, Xmax, numbergrids).tolist())  
    
for i in np.arange(int(maturity/delta_t)):
    X.append([i/int(maturity/delta_t)]*delta_x)

Z = np.array([el.tolist() for el in matrixsol])

print(len(X), len(X[0]))
print(len(Y), len(Y[0]))
print(len(Z), len(Z[0]))

print(matrixsol[-1])
print(matrixsol[-2])
```
